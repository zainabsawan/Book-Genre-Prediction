```python
import pandas as pd
```

```python
# Loading csv
df = pd.read_csv("../data/GoodReads_books_cleaned.csv")
```

```python
# Using set() to get unique values
# iterating through nested for loop - g in genre, genre in sublist, sublist in df
all_genres = set([g for sublist in df["genre"] for g in sublist.split(",")])

#converting back to list for ____
all_genres = list(all_genres)
print(all_genres)
```

```python
from sentence_transformers import SentenceTransformer
# import tqdm

# Loading a small but strong model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Encoding genres into embeddings
embeddings = model.encode(all_genres, show_progress_bar = True)
```

```python
from sklearn.cluster import KMeans

# 15 broad genres
num_clusters = 15

# random_state for reproducibility
clustering_model = KMeans(n_clusters=num_clusters, random_state=42) 

clustering_model.fit(embeddings)

cluster_assignment = clustering_model.labels_
```

```python
# Inspecting clusters

clustered_genres = pd.DataFrame({"genre": all_genres, "cluster": cluster_assignment})

# viewing results
for cluster_id in range(num_clusters):
    print(f"\nCluster {cluster_id}")
    print(clustered_genres[clustered_genres["cluster"] == cluster_id]["genre"].values)
```
